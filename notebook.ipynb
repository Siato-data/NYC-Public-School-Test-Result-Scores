Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

```

1. School with the best math scores 

```python
#subset the data with 80% score in math 
best_math_schools= schools[schools['average_math']>= 640]

#keep only columns 'school_name' and 'average_math' in a descending way 
best_math_schools= best_math_schools[['school_name','average_math']].sort_values('average_math',ascending=False)

#print the result
print(best_math_schools.head())
```

2. Top 10 schools performing

```python
# creat column 'total_SAT' for total score
schools['total_SAT']= schools['average_math']+schools['average_reading']+schools['average_writing']

# Show in dataframe column 'total_SAT' in descending order
schools=schools.sort_values('total_SAT', ascending=False)

# the 10 best score in DataFrame top_10_schools columns 'school_name' an 'total_SAT'
top_10_schools=schools[['school_name','total_SAT']].head(10)

print(top_10_schools)
```

3. New York Borough with higher standard deviation

```python
#mean, standard deviation and number of school of each borough 
largest_std_dev =schools.pivot_table(index='borough', aggfunc={'total_SAT':['count','mean','std']}).round(2)

#rename the columns mean, standard deviation and number of school
largest_std_dev.columns=['num_schools','average_SAT','std_SAT']

#boroughs who got the highest score (only one here)
largest_std_dev= largest_std_dev.loc[largest_std_dev['std_SAT']==largest_std_dev['std_SAT'].max()] 

print(largest_std_dev)
```
